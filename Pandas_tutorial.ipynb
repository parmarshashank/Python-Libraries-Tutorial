```python
# DataFrame¶
# A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.
```

```python
pd.DataFrame({"col1": [1, 2, 3, 4], "col2": [3,4,5,6]} )  #They must be ordered for ex: it is not possible to make a dataframe using set as sets in python are unordered
```
