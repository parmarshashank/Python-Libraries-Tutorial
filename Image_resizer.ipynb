```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
path="./Agent P (Perry the Platypus).jpg"
```

```python
image=Image.open(path)
img_arr=np.array(image)
```

```python
plt.imshow(img_arr)
plt.title("Agent P")
plt.axis('off')
plt.show()
```

```python
print(img_arr)
img_arr.shape
```

```python
def resize_image(image_array, new_size):
    original_height, original_width = image_array.shape[:2]
    new_height, new_width = new_size
    resized_image = np.zeros((new_height, new_width, image_array.shape[2]), dtype=image_array.dtype)
    row_ratio = original_height / new_height
    col_ratio = original_width / new_width

    for i in range(new_height):
        for j in range(new_width):
            original_i = int(i * row_ratio)
            original_j = int(j * col_ratio)
            resized_image[i, j] = image_array[original_i, original_j]

    return resized_image
```

```python
newSize=(200, 200)
new_img= resize_image(img_arr, newSize)
```

```python
plt.imshow(new_img)
plt.title("Agent P after resizing")
plt.axis('off')
plt.show()
```
